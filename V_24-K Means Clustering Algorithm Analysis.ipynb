### What is K-Means Clustering?
- k-means performs division of objects into clusters which are "similar" between them and are "disimilar" to the objects belonging to another cluster
- type of unsupervised machine learnig

### How does K-Means Clustering work?
![](pic/v24/11s.png)

##### 1. Elbow pont
- let's say we have a dataset for Grocery shop
![](pic/v24/12s.png)

- the important question is: how would you choose the optimum number of clusters?
- The best way to do this is by <b>Elbow Method</b>
- The idea of the elbow method is to run K-Means clustering on the dataset where "k" is reffered as number of clusters
- Within <b>sum of squared (WSS)</b> is defined as the sum of the squared distance between each member of the cluster and its centroid
![](pic/v24/13s.png)

- where Xi is the data point and Ci is the closest point to the centroid
- now, we draw a curve between WSS and the number of clusters
![](pic/v24/14s.png)

- here, we can see a very slow change in the value of WSS after k=2, so you should take that elbow point value as the final number of clusters

##### 2. Measure the Distance
- the given data points below are assumed as delivery points
![](pic/v24/15s.png)

- we can randomly initialize two points called the cluster centroid, <b>Euclidean distance</b> is a distance measure used to find out which data poin is closest to our centroids
![](pic/v24/16s.png)

##### 3. Grouping
- based upon the distance from c1 and c2 centroids, the data points will group itself into clusters
![](pic/v24/17s.png)

##### 4. Reposition the centroids
- compute the centroid of data point inside the blue cluster
- reposition the centroid of the blue cluster to the new centroid
![](pic/v24/18s.png)

- now, compute the centroid of data points inside the orange cluster
- reposition the centroid of the orange cluster to the new centroid
![](pic/v24/19s.png)

##### 5. Convergence
- once the clusters become static, K-Means clustering algorithm is said to be converged
![](pic/v24/20s.png)

### K Means Clustering Algorithm - calculation
Assuming we have inputs x1, x2, x3, .... and a value of k

- Step 1: Pick k random points as cluster centers called centroids
- Step 2: Assign each xi to the nearest cluster by calculating its distance to each centroid
- Step 3: Find new cluster center by taking the average of the assigned points
- Step 4: Repeat Step 2 and 3 until none of the cluster assignents change

![](pic/v24/21s.png)

## K Means Clustering - demo example
***

#### Problem Statment:
Walmart wants to open a chain of stores across Florida and wants to find out optimal store to locations to maximize revenue.

#### Solutions
- Walmart already has a strong e-commerce presence.

- Walmart can use its online customer data to analyze the customer locations along with the monthly sales.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
```

```python
from sklearn.datasets.samples_generator import make_blobs

X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_means = kmeans.predict(X)
```

```python
y_means # number of cluster for obsevations - 300 samples
```

```python
from sklearn.metrics import pairwise_distances_argmin

def find_clusters(X, n_clusters, rseed=2):
    # 1. randomly choose clusters
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
    centers = X[i]
    
    while True:
        # 2. assign labels based on closest center
        labels = pairwise_distances_argmin(X, centers)
        
        # 3. find new centers from means of points
        new_centers = np.array([X[labels == i].mean(0) for i in range(n_clusters)])
        
        # 4. check for convergence
        if np.all(centers == new_centers):
            break
        centers = new_centers
    return centers, labels

centers, labels = find_clusters(X, 4)
plt.scatter(X[:, 0], X[:, 1], c=y_means, s=50, cmap="viridis")

plt.scatter(centers[:, 0], centers[:, 1], c="red", s=200, alpha=0.5)
# red are the centers of the clasters
```

# 2. K Means Clustering Algorithm - practical example
***

### Problem stament
- to perform color compression on images using K-Means algorithm

```python
from sklearn.datasets import load_sample_image
china = load_sample_image("flower.jpg")
ax = plt.axes(xticks=[], yticks=[])
ax.imshow(china);
```

```python
china.shape
```

```python
china
```

```python
data = china / 255.0 # use 0...1 scale
data = data.reshape(427 * 640, 3)
data.shape
```

```python
data
```

```python
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data
    
    # choose a random subset
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = data[i].T
    
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=20);
```

```python
plot_pixels(data, title='Input color space: 16 million possible colors')
```

```python
import warnings; warnings.simplefilter('ignore')  # Fix NumPy issues.

from sklearn.cluster import MiniBatchKMeans
kmeans = MiniBatchKMeans(16)
kmeans.fit(data)
new_colors = kmeans.cluster_centers_[kmeans.predict(data)]

plot_pixels(data, colors=new_colors,
            title="Reduced color space: 16 colors")
```

```python
china_recolored = new_colors.reshape(china.shape)

fig, ax = plt.subplots(1, 2, figsize=(16, 6),
                       subplot_kw=dict(xticks=[], yticks=[]))
fig.subplots_adjust(wspace=0.05)
ax[0].imshow(china)
ax[0].set_title('Original Image', size=16)
ax[1].imshow(china_recolored)
ax[1].set_title('16-color Image', size=16);
```
